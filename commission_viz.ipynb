# Commission Costs

Commision cost that I have to take into accout:

- trading fee (0.1% on nordics stocks, 0.2% on internation stocks, min 29DKK)
- spread 
- custody costs (varies from security to security)
- taxes 

```python
import numpy as np
import matplotlib.pyplot as plt
```

Until I won't make a transaction bigger than 29.000 DKK (approx 3981â‚¬) I'll pay 29 DKK per transaction, so given a $x$ number of different stocks I'll pay $x \times 29$ DKK of commision

```python
# Plot commission based on the number of stocks
# y = 29 * x
x = np.arange(0, 15, 1)
y = 29 * x

fig = plt.figure(figsize=(10, 5))

plt.plot(x, y)
plt.title('Fee with the number of shares purchased')
plt.ylabel('DKK')
plt.xlabel('number of stocks')
plt.grid()
plt.show()
```

```python
x1 = np.linspace(1000, 40000)
f1 = lambda x: 0.2 * x if x < 29000 else (0.1 * x) + 2900
vf1 = np.vectorize(f1)

plt.plot(x1, vf1(x1))
plt.axvline(x=29_000, color='red')
plt.xlabel('Amount (DKK)')
plt.ylabel('Commission')
plt.grid()

plt.show()
```
